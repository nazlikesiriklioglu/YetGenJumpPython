### Soru 4
x ve y a**```python
dında iki liste oluşturunuz. Bu listeleri birbiryle yer değiştirerek yazdırınız.
```**

```python
x = [1,2,3,4,5]
y = [5,4,3,2,1]
``` 

```python	
Output:
x = [5, 4, 3, 2, 1]
y = [1, 2, 3, 4, 5]
```



````python
x=[1,2,3,4,5]
y=[5,4,3,2,1]

# listeleri değiştirmemiz için üçüncü bir değişkene ihtiyacımız var. Bu sebeple ven z değişkenini kullandım.

z=x 
x=y # Bu atamayla x değişkeni y değişkeni haline geldi.
y=z # Artık y değişkeninin değeri z değişkenine eşit olacak. Yani y değişkeni x değişkeni haline geldi.
print("x=", x)
print("y=", y) 
````
