# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def urunEkle():
    with open("urunler.txt", "a") as dosya:
        urunSayisi = int(input("Girmek istediğiniz ürün sayısını giriniz:"))
        for i in range(urunSayisi):
            isim = input("Ürünün ismini giriniz:")
            fiyat = int(input("Ürünün fiyatını giriniz:"))
            dosya.write(f"{isim},{fiyat}\n")
    with open("urunler.txt", "r") as dosya:
        veriler = dosya.read()
        print(veriler)
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class Vehicle:
    def __init__(self,name,speed,mileage):
        self.name=name
        self.speed=speed
        self.mileage=mileage
class Bus(Vehicle):
    pass
bus=Bus("School Volvo",180,12)
print(f"Name={bus.name}\nSpeed={bus.speed}\nMileage={bus.mileage}\n")
````
