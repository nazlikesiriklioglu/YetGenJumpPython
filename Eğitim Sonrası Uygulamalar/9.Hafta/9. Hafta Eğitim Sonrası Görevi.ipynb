# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class kumanda:
    def __init__(self,tvDurum="Kapalı",tvSes=0,kanalListesi=["Fox"],kanal="Fox"):
        self.tvDurum= tvDurum
        self.tvSes=tvSes
        self.kanalListesi=kanalListesi
        self.kanal=kanal
    
    def tvKapat(self):
        if self.tvDurum=="Açık":
           print("Tv kapatılıyor.")
           self.tvDurum = "Kapalı"

    def tvAc(self):
        if self.tvDurum=="Kapalı":
           print("Tv Açılıyor.")
           self.tvDurum = "Açık"
            
    def sesAyarlama(self):
        while 1>0:
            i=input("<:sesi azalt,>: sesi yükselt anlamlarına gelmektedir:")
            if (i == "<"):
                if (self.tvSes!= 0):
                    self.tvSes-= 1
                    print("Ses:",self.tvSes)
            elif (i == ">"):
                if (self.tvSes!= 50):
                    self.tvSes+= 1
                    print("Ses:",self.tvSes)
            else:
                print("Ses Güncellendi:",self.tvSes)

    def kanalEkle(self,kanalİsimleri):
        print("Kanal ekleniyor...")
        time.sleep(2)
        self.kanalListesi.append(kanalİsimleri)
        print("Kanal eklendi")

    def rastgeleGecis(self):
        a=random.randint(0,len(self.kanalListesi)-1)
        
        self.kanal = self.kanalListesi[a]
        print("Şu anki Kanal:", self.kanal)

    def __str__(self):
         return "Tv Durumu : {}\nSes: {}\nKanallar: {}\nŞu anki kanal: {}\n".format(self.tvDurum,self.tvSes,self.kanalListesi,self.kanal)
    def __len__(self):
        return  len(self.kanalListesi)

    print("1. Tv Aç\n2. Tv Kapat\n3. Ses Ayarları\n4. Kanal Ekle\n5. Kanal Sayısını Öğrenme\n6. Rastgele Kanala Geçme\n7. Televizyon Bilgileri\nÇıkmak için q ya basınız")
    kumanda= kumanda()   
     

    def islemler(self):
        while 1>0:
            islem=input("Yapılacak işleme atanan sayiyi giriniz:")

            if (islem=="1"):
                kumanda.tvAc()

            elif (islem=="2"):
                kumanda.tvKapat()
        
            elif (islem=="3"):
                kumanda.sesAyarlama()

            elif (islem=="4"):
                kanalİsimleri=input("Eklemek istediğiniz kanalların isimlerini virgül ',' ile ayırarak giriniz:")
                kanalListesi=kanalİsimleri.split(",")
                for i in kanalListesi:
                    kumanda.kanalEkle(i)
        
            elif (islem=="5"):
                print("Kanal sayisi=",len(kanalListesi))
        
            elif (islem=="6"):
                kumanda.rastgeleGecis()

            elif (islem=="7"):
                print(kumanda)
        
            elif (islem=="q"):
                print("Çıkış yapılıyor...")
        
        
```
